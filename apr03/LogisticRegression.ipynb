## Logistic Regression
Logistic regression is a statistical method used to analyze data in order to predict a binary outcome, such as whether a customer will make a purchase or not. However, when we have more than two possible outcomes, we need to use a multinomial logistic regression algorithm.

Multinomial logistic regression is an extension of binary logistic regression, which is used when the dependent variable has more than two categories. This algorithm models the probability of each category of the dependent variable, given the values of the independent variables. The output of the algorithm is a set of coefficients for each independent variable, which can be used to calculate the probability of each possible outcome.

In multinomial logistic regression, the dependent variable has more than two possible outcomes, which are also known as classes. For example, if we are trying to predict a person's profession based on their age, education, and gender, we might have three classes: doctor, lawyer, and engineer. The algorithm then estimates the probability of each class for a given set of input values.

One important aspect of multinomial logistic regression is that it uses a set of reference categories. One of the categories is chosen as the reference category, and the coefficients for the other categories are estimated relative to the reference category. This means that the output of the algorithm provides information about the probability of each class relative to the reference category.

Here we will use it on the Iris dataset to predict the species of the flower based on the sepal and petal length and width. SciKit Learns implementaion of Logistic Regression is looking at the target variable and seeing if it is binary or not. If it is binary, it will use the Logistic Regression algorithm. If it is not binary, it will use the Multinomial Logistic Regression algorithm. As we are using the Iris dataset, which has 3 species, it will use the Multinomial Logistic Regression algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

```python
# Load the Iris dataset
iris = load_iris()
```

```python
# Use the first two features, sepal length and sepal width as X
X = iris.data[:, :2]
```

```python
X
```

```python
# Get the target labels
y = iris.target
y
```

```python
clf_lr = LogisticRegression()
clf_lr.fit(X, y)
```

```python
clf_lr.predict([[6.0, 3.0]])
```

```python
class_names = iris.target_names
class_names[clf_lr.predict([[2.7, 1.1]])[0]]
```

```python
# Plot the data points and decision boundary
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = clf_lr.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.bwr, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, alpha=0.5)

# Add labels and title
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Logistic Regression on Iris Dataset')

# Show the plot
plt.show()
```

The three colored fields represent the three different classes in the iris dataset (Setosa in blue, Versicolor in red, and Virginica in light blue). The dots represent the input features (sepal length and sepal width) of each sample in the dataset. The color intensity of the dots corresponds to the probability of each sample belonging to the Setosa class.

The background color represents the decision boundary of the logistic regression classifier. Points that are closer to the blue region have a higher probability of belonging to the Setosa class, while points that are closer to the red or light blue regions have a lower probability of belonging to the Setosa class. The decision boundary is determined by the logistic regression algorithm and is a curve because the classes are not completely separable with a straight line.

The goal of logistic regression is to learn a decision boundary that can separate the two classes by mapping the input space to a predicted probability space. The decision boundary separates the two classes based on the probability of each input belonging to the positive (Setosa) or negative (not Setosa) class.
