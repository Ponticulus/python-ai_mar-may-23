## Linear Regression
Linear regression is a linear approach to modeling the relationship between a scalar response and one or more explanatory variables (also known as dependent and independent variables). 

The dependent variable is the output or the predicted value that we are interested in. It is also called the target variable or response variable. The dependent variable is the variable that is being predicted or estimated by the machine learning model.

The independent variables, also known as the features or predictors, are the input variables used to make predictions. These variables are used to provide information to the model so that it can make predictions on the dependent variable. Independent variables can be numeric, categorical or ordinal.

For example, let's say we want to build a machine learning model that predicts the price of a house. In this case, the price of the house would be the dependent variable. The independent variables could include things like the size of the house, the number of bedrooms, the location of the house, etc. These variables would be used to predict the price of the house.

The case of one explanatory variable is called simple linear regression. For more than one explanatory variable, the process is called multiple linear regression. This section covers simple linear regression.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Create a noisy dataset
np.random.seed(42)
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 2 * X + 1 + np.random.randn(50, 1)
```

```python
# Create a linear regression model
lr = LinearRegression()
lr.fit(X, y)
```

```python
# Plot the dataset
plt.scatter(X, y, color='blue')
plt.plot(X, lr.predict(X), color='red')

# Show the plot
plt.show()
```

Nonlinear relationships are also common. Here we look at simple example of a 2nd order polynomial.

```python
# Create a noisy dataset
np.random.seed(42)
X = np.linspace(-1, 1, 50).reshape(-1, 1)
y = 0.5 * X ** 2 + X + np.random.randn(50, 1) * 0.1
```

```python
# Transform the data to include polynomial terms up to the degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
# Create a linear regression model
lr = LinearRegression()
lr.fit(X_poly, y)
```

```python
# Plot the dataset
plt.scatter(X, y, color='blue')
plt.plot(X, lr.predict(poly.fit_transform(X)), color='red')

# Show the plot
plt.show()
```

Work with the iris dataset. This is a classic dataset in machine learning and statistics. It is included in scikit-learn in the datasets module. It contains 150 observations of iris flowers. The data contains the sepal and petal length and width measurements for each flower. The species of each flower is also recorded. The dataset is often used for testing out machine learning algorithms and visualizations (for example, scatter plots).

```python
from sklearn.datasets import load_iris
```

```python
# Load the iris dataset
iris = load_iris()
```

```python
# Print it
iris
```

```python
# As we work with a regression algorithm we will predict the petal width from the petal length
X = iris.data[:, 2].reshape(-1, 1) # Petal length
y = iris.data[:, 3] # Petal width
```

```python
clf_lr = LinearRegression()
clf_lr.fit(X, y)
```

```python
y_pred = clf_lr.predict(X)
```

```python
y_pred
```

```python
# Plot the result
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Linear Regression on Iris Dataset')
plt.show()
```
